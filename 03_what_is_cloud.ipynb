<h3 style='font-size:30px;font-style:italic'> O problema dos Data Centers Locais</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Os donos do website - ou qualquer outro serviço - que é hospedado em um servidor local precisam lidar com uma série de problemas/riscos.
        </li>
        <li>
            Poderíamos mencionar os custos de manutenção, necessidade de uma equipe disponível 24/7 e a possibilidade de desastres na localidade do data center.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> What is Cloud Computing?</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            A Computação em Nuvem surgiu como uma solução às questões mencionadas. Essa consiste na utilização de um servidor remoto pertencente a um terceiro.
        </li>
        <li> 
            À medida que nossa demanda por processamento ou armazenamento aumenta, podemos requisitar o aumento ou troca de nossos servidores por outros instantaneamente. Dessa forma, nos certificamos de que sempre teremos um poder computacional em linha com nossas necessidades. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Modalidades de Nuvem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            <i> Nuvem Privada</i>: É um serviço de nuvem exclusivo a uma organização e não acessível fora dela. Isso não impede que os servidores sejam geridos por uma empresa terceira (ex: Rackspace).
            <p style='margin-left:142px'> 
                Garante maior controle e segurança de dados sensíveis.
            </p>
        </li>
        <li>
            <i> Nuvem Pública</i>: É o fornecimento de poder computacional pertencente a um terceiro via Internet. 
        </li>
        <li> 
            <i> Nuvem Híbrida</i>: É o uso simultâneo de um serviço de nuvem privada com um de núvem pública, garantindo segurança a dados sensíveis e flexibilidade computacional.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Características do Cloud Computing</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            <i> Auto-Serviço Sob Demanda</i>: Os clientes podem obter ou descartar recursos por conta própria, sem a necessidade de interação com um funcionário da AWS.
        </li>
        <li> 
            <i> Amplo Acesso a Recursos</i>: Os recursos disponíveis na rede da AWS podem ser consumidos de diversas maneiras.
        </li>
        <li> 
            <i> Multi-Posse e Compartilhamento de Recursos</i>: Todos os clientes do provedor usufruirão da mesma infraestrutura simultaneamente.
        </li>
        <li> 
            <i> Elasticidade e Escalabilidade</i>: A obtenção ou descarte de recursos acontece de maneira instantânea. Isso pode acontecer tanto por desejo do usuário, quanto via monitoramento da demanda por parte da AWS.
        </li>
        <li> 
            <i> Serviço sob Medida</i>: Pague apenas pelo poder computacional utilizado.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Six Advantages of Cloud Computing</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            <i> CAPEX por OPEX</i>: O Capital Expenditure será reduzido por conta do menor uso de servidores próprios. Por sua vez, o OPEX tende a crescer por conta das tarifas de aluguel da AWS.
        </li>
        <li> 
            <i> Se Beneficie das Economias de Escala</i>: A provisão de computação sob demanda oferece a otimização de custos à empresa.
        </li>
        <li> 
            <i> Parar de Estimar Capacidade Necessária</i>: A Escalabilidade da AWS torna dispensável a necessidade de cálculos de demandas futuras.
        </li>
        <li> 
            <i> Mais Velocidade e Agilidade</i>: A possibilidade de auto-serviço e escalabilidade acelera e agiliza a produtividade dos seus consumidores.
        </li>
        <li> 
            <i> Gastos zero com Data Centers</i>: Uma vez terceirizando a provisão de servidores, a posse de um data center torna-se desnecessária.
        </li>
        <li> 
            <i> Torne-se Global em Minutos</i>: Aproveite a presença mundial da AWS a fim de disponibilizar os serviços da sua empresa internacionalmente.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Types of Cloud Computing</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Além dos diferentes graus de acessibilidade, os serviços de nuvem providenciam ao usuário níveis distintos de controle sobre o serviço prestado. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Iaas</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            É a modalidade mais flexível de tipo de serviços. Nela, o provedor se compromete a cuidar de toda a infraestrutura do data center - servidores, rede, armazenamento -, enquanto que o cliente tem completa liberdade de gerir o middleware, sistema operacional e dados tidos.
        </li>
        <li> 
            Na AWS, o EC2 é um serviço designado a essa modalidade de nuvem.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> PaaS</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em PaaS, o provedor da nuvem se responsabiliza não apenas pela infra, como também por outros componentes primários das aplicações, como OS e middleware. O usuário gerencia apenas os apps e dados presentes no servidor.
        </li>
        <li>
            O Elastic Beanstalk é um serviço PaaS fornecido pela Amazon.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> SaaS</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Em SaaS, o provedor de cloud é responsável por gerir todo o produto desenvolvido pelos seus usuários. Isso significa que, além da parte física, de OS e middleware, a AWS gere os dados e o software criado.
        </li>
        <li> 
            Sites e aplicativos mobile costumam estar empregados nos servidores na modalidade SaaS.
        </li>
        <li> 
            O Rekognition é um dos vários serviços SaaS fornecidos pela AWS.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Pricing of the Cloud</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            A AWS impõe custos a seus usuários de três maneiras diferentes.
            <ul> 
                <li> 
                    Pelo tempo de computação.
                </li>
                <li> 
                    Pelos dados armazenados nos servidores.
                </li>
                <li>
                    Pelos dados que saem da nuvem; tudo o que entra no servidor não é cobrado.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> AWS Cloud Overview</h2>
<h3 style='font-size:30px;font-style:italic'> AWS Global Infrastructure</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            A Amazon impõe uma hierarquia à sua infraestrutura. É bastante importante se situar quanto aos termos relativos a essa ordenação.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Regions</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Consiste em um conjunto de data centers.
        </li>
        <li> 
            A escolha da região ideal depende de uma série de questões relativas ao negócio da sua empresa.
            <ul> 
                <li> 
                    <i> Proximidade com Consumidores:</i> Os data centers devem se encontrar geograficamente próximos dos principais clientes de sua aplicação a fim de evitar problemas de latência. 
                </li>
                <li> 
                    <i> Compliance:</i> Legislações de governos acerca do uso de dados têm influência quanto à escolha dos data centers. Um certo país pode afirmar que informações de seus habitantes devem ser armazenadas dentro de seu território. 
                </li>
                <li> 
                    <i> Disponibilidade de Serviços:</i> Nem todas as regiões possuem a mesma oferta de serviços que outras; tenha certeza de que o cluster que você escolheu oferece todos os componentes necessários à sua aplicação.
                </li>
                <li> 
                    <i> Pricing:</i> Cada região tem um sistema de precificação distinto. Leia com atenção os termos do contrato!
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Availability Zones</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            As zonas de disponibilidade constituem uma região. Elas são um ou mais data centers distintos com energia, rede e conectividade redundantes. 
        </li>
        <li> 
            As AZ's estão isoladas geograficamente entre si de maneira a evitar que desastres em uma delas acarretam em problemas em outras. 
        </li>
        <li> 
            Todas as zonas de disponibilidade de uma certa região são ligadas por uma rede de alta banda-larga e baixa latência.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Points of Presence</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            A AWS possui mais de 400 pontos de presença ao redor do mundo. Isso permite que os serviços sejam oferecidos com baixa latência aos seus usuários.
        </li>
    </ul>
</div>
