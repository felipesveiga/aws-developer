<h1 style='font-size:40px'> CloudFront</h1>

<h2 style='font-size:30px'> CloudFront - Overview
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O CloudFront é um serviço que busca armazenar em cache por diversas partes do mundo dados frequentemente requisitados.
        </li>
        <li>
            Seu propóstio é garantir que sua aplicação retorne seu conteúdo o mais rápido possível independentemente de onde seu cliente estiver.
        </li>
        <li>
            Os dados ficam armazenados em Edge Locations da AWS.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
            <img src='img/15_edge_locations.png'>
        <figcaption style='font-size:15px'>
            Mapa das Edge Locations pelo globo.
        </figcaption>
    </center>
</figure>

<h3 style='font-size:30px;font-style:italic'> Origins</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Definimos como origem o lugar de onde o recurso a ser cacheado virá. No caso do CloudFront, essa pode ser um bucket S3, uma VPC ou qualquer backend HTTP.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
            <img src='img/15_cloudfront_diagram.png'>
        <figcaption style='font-size:15px'>
            Funcionamento do CloudFront.
        </figcaption>
    </center>
</figure>

<h3 style='font-size:30px;font-style:italic'> CloudFront vs S3 Cross Region Replication</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            O CloudFront é um serviço de disponibilização global de seus conteúdos. Nele, os arquivos são armazenados em cache, respeitando um TTL.
        </li>
        <li>
            Já o S3 Cross Region Replication é um serviço de escopo regional, em que os arquivos são atulizados near real-time.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> 
    CloudFront - Caching & Caching Policies
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Todo o objeto armazenado no cache do CloudFront receberá um identificador.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> 
    CloudFront - Cache Invalidations
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Um problema muito comum em aplicações com cache é a defasagem de informações. Caso o registro já tido no cache seja atualizado, os usuários conseguirão apenas enxergar a versão antiga.
        </li>
        <li>
            Para evitar isso, podemos criar um mecanismo que apague dados do cache assim que eles sofram alguma atualização no backend.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> 
    CloudFront - Cache Behaviors
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O CloudFront também suporta customizarmos o seu comportamento, com base nas particularidades de cada requsição do usuário. 
        </li>
        <li>
            Conseguimos, inclusive, permitir ou negar acesso ao conteúdo com base na localização do usuário.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> 
    CloudFront Signed URL / Cookies
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O CloudFront possui capacidade de gerar pre-signed URL's ou cookies que garantem acesso a recursos específicos em aplicações disponibilizadas na AWS.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Múltiplas Edge Locations</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A fim de garantir a menor latência possível para seus clientes, talvez valha considerar a replicação dos seus dados em múltiplas edge locations.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Multiple Origin</h3>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Como vimos anteriormente, podemos definir diferentes pontos de origem dos dados, a depender da rota requisitada pelo cliente.
        </li>
    </ul>
</div>

<div> 
    <ul style='font-size:20px'> 
        <li>
            A fim de garantir mais segurança aos dados de nossos clientes, é extremamente recomendado o uso do protocolo HTTPS em suas aplicações.
        </li>
        <li>
            Por estar sempre a frente dos servidores web, recomenda-se depositar a chave pública na edge location, para que o dado seja posteriormente decodificado na aplicação via chave privada.
